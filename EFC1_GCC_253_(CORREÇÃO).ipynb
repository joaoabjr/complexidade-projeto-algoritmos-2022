# Funções de Ordenação



##  Insertion Sort


```python
def insertionSort(array):
  comparacao = 0
  for j in range (1, len(array)):
    comparacao += 1
    chave = array[j]
    i = j - 1
    while i >= 0 and array[i] > chave:
      comparacao += 1
      array[i + 1] = array[i]
      i = i - 1
    comparacao += 1
    array[i + 1] = chave
  comparacao += 1
  return array, comparacao
```

##  Merge Sort


```python
def mergeSort(array):
  comparacao = 0
  if len(array) > 1:
    comparacao += 1
    meio = int(len(array)/2)
    i, j, k = 0, 0, 0
    arrayEsquerda = array[:meio] # até o meio
    arrayDireita = array[meio:] # do meio para frente

    mergeSort(arrayEsquerda)
    mergeSort(arrayDireita)

    while i < len(arrayEsquerda) and j < len(arrayDireita):
      comparacao += 1
      comparacao += 1
      if arrayEsquerda[i] < arrayDireita[j]:
        array[k] = arrayEsquerda[i]
        i = i + 1
      else:
        array[k] = arrayDireita[j]
        j = j + 1
      k = k + 1 #iterando com o K
      comparacao += 1
    comparacao += 1
    while i < len(arrayEsquerda):
      comparacao += 1
      array[k] = arrayEsquerda[i]
      i = i + 1
      k = k + 1
    comparacao += 1
    while j < len(arrayDireita):
      comparacao += 1
      array[k] = arrayDireita[j]
      j = j + 1
      k = k + 1
    comparacao += 1
  comparacao += 1
  return array, comparacao
```

##  Selection Sort


```python
def selectionSort(array):
  comparacao = 0
  for i in range(len(array)):
    comparacao += 1
    menor = i
    for j in range(i + 1, len(array)):
        comparacao += 1
        if array[j] < array[menor]:
          comparacao += 1
          menor = j

    if array[i] != array[menor]:
        comparacao += 1
        aux = array[i]
        array[i] = array[menor]
        array[menor] = aux  
  return array, comparacao
```

##  Bubble Sort 


```python
def bubbleSort(array):
  comparacao = 0
  elementos = len(array)-1
  for i in range(0, elementos):
    comparacao += 1
    for j in range(0, elementos):
      comparacao += 1
      if array[j] > array[j + 1]:
        comparacao += 1
        array[j], array[j + 1] = array[j + 1], array[j]
  return array, comparacao
```

## Heap Sort

```python
def heapSort(array):
  comparacao = 0
  tamanho = len(array)
  # transformando em MaxHeap
  for i in range(tamanho//2 - 1, -1, -1):
    comparacao += 1
    heapify(array, tamanho, i)

  for i in range(tamanho - 1, 0, -1):
    comparacao += 1
    array[i], array[0] = array[0], array[i]
    heapify(array, i, 0)
  return array, comparacao

def heapify(array, tamanho, i):
  comparacao = 0
  max = i
  filhoEsquerdo = 2 * i + 1
  filhoDireito = 2 * i + 2
  if filhoEsquerdo < tamanho and array[i] < array[filhoEsquerdo]:
    comparacao += 1
    max = filhoEsquerdo
  if filhoDireito < tamanho and array[max] < array[filhoDireito]:
    comparacao += 1
    max = filhoDireito
  if max != i:
    comparacao += 1
    array[i], array[max] = array[max], array[i]
    heapify(array, tamanho, max)
```

## Quick Sort

> pivo = *A[A-comprimento]*

```python
def quickSort(array, inicio, fim):
  comparacao = 0

  comparacao += 1
  if inicio > fim:
    return comparacao

  ptrA = inicio
  ptrP = fim
  pivo = array[-1]

  while ptrA < ptrP:
    comparacao += 1
    while ptrA < ptrP and array[ptrP] > pivo:
      comparacao += 1
      ptrP = ptrP - 1
    comparacao += 1

    comparacao += 1
    if ptrA < ptrP:
      array[ptrA] = array[ptrP]
      ptrA = ptrA + 1

    while ptrA < ptrP and array[ptrA] <= pivo:
      comparacao += 1
      ptrA = ptrA + 1
    comparacao += 1
    
    comparacao += 1
    if ptrA < ptrP:
      array[ptrP] = array[ptrA]
      ptrP = ptrP - 1

    array[ptrA] = pivo
  comparacao += quickSort(array, inicio, ptrA - 1)
  comparacao += quickSort(array, ptrA + 1, fim)
  return comparacao

def quickSortArray(array):
  return quickSort(array, 0, len(array) - 1)
```

## 2. Desenvolva uma versão aleatorizada do QUICKSORT (pivo aleatório) e compare com a versão apresentada em sala de aula e definida acima.

```python
def quickSortPivoAleatorio(array, inicio, fim):
  comparacao = 0

  comparacao += 1
  if inicio > fim:
    return comparacao

  ptrA = inicio
  ptrP = fim
  pivo = np.random.randint(inicio, fim + 1)

  while ptrA < ptrP:
    comparacao += 1
    while ptrA < ptrP and array[ptrP] > pivo:
      comparacao += 1
      ptrP = ptrP - 1
    comparacao += 1

    comparacao += 1
    if ptrA < ptrP:
      array[ptrA] = array[ptrP]
      ptrA = ptrA + 1

    while ptrA < ptrP and array[ptrA] <= pivo:
      comparacao += 1
      ptrA = ptrA + 1
    comparacao += 1
    
    comparacao += 1
    if ptrA < ptrP:
      array[ptrP] = array[ptrA]
      ptrP = ptrP - 1

    array[ptrA] = pivo
  comparacao += quickSortPivoAleatorio(array, inicio, ptrA - 1)
  comparacao += quickSortPivoAleatorio(array, ptrA + 1, fim)
  return comparacao

def quickSortArray(array):
  return quickSortPivoAleatorio(array, 0, len(array) - 1)
```

```python
melhorCasoPivotagem, piorCasoPivotagem, casoAleatorioPivotagem = {}, {}, {}
melhorCaso, piorCaso, casoAleatorio = {}, {}, {}

for i in range(5, 1000, 5):
  arrayAleatorio = np.random.randint(0, 500000, i)
  arrayOrdenado = np.sort(arrayAleatorio)
  arrayInverso = arrayOrdenado[::-1]
  casoAleatorioPivotagem[i] = quickSortPivoAleatorio(arrayAleatorio, 0, len(arrayAleatorio) - 1)
  melhorCasoPivotagem[i] = quickSortPivoAleatorio(arrayOrdenado, 0, len(arrayAleatorio) - 1)
  piorCasoPivotagem[i] = quickSortPivoAleatorio(arrayInverso, 0, len(arrayAleatorio) - 1)
  casoAleatorio[i] = quickSortArray(arrayAleatorio)
  melhorCaso[i] = quickSortArray(arrayOrdenado)
  piorCaso[i] = quickSortArray(arrayInverso)

plt.plot(list(piorCasoPivotagem.keys()), list(piorCasoPivotagem.values()), label='Pior caso')
plt.plot(list(casoAleatorioPivotagem.keys()), list(casoAleatorioPivotagem.values()), label='Caso Aleatório')
plt.plot(list(melhorCasoPivotagem.keys()), list(melhorCasoPivotagem.values()), label='Melhor caso')
plt.title("Quick Sort Pivô Aleatório")
plt.legend(title="Legenda")
plt.xlabel("Tamanho")
plt.ylabel("Quantidade comparações")
plt.show()
print()
plt.title("Quick Sort")
plotagem(quickSortArray) # QuickSort versão da letra a)
```
